# Face Detection using opencv

```python
# haarcascade file download code

import urllib.request

# URL of Haar Cascade XML file for face detection
url = 'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'

# Save location â€” you can save it in your current directory
filename = 'haarcascade_frontalface_default.xml'

# Download the file
urllib.request.urlretrieve(url, filename)

print("Haar Cascade XML file downloaded successfully!")
```

```python
import cv2
from google.colab.patches import cv2_imshow # 1. comment it if in jupyter
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

img = cv2.imread(r"/content/img5.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
cv2_imshow(img)  # 2. for jupyter cv.imshow('Detected Face', img)  imgname, img
cv2.waitKey(0)
cv2.destroyAllWindows()
```
